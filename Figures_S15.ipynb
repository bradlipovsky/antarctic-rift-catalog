```python
ftsz=10

fig = plt.subplots(3,2,figsize=(15,15),sharey=True)

#-----------------------------------------------------
ax1=plt.subplot(3,2,1)
   
rift_len = 40
xx = np.arange(0,rift_len,0.01)
w1 = 254.5 #rate m/yr
r = np.sqrt(rift_len**2 - xx**2)

n=1

pl_t1 = ax1.plot(xx,(w1 * ((r/rift_len)**n)),'-',linewidth=2,color=(1,2/3,2/3),label='1 year')
    
ax1.plot([0,0],[0,w1],'--',color='0.5',linewidth=2)
ax1.plot([1,1],[0,(w1 * ((r/rift_len)**n))[100]],'--',color='0.5',linewidth=2)
    
#test fit at GNSS
ax1.plot([16,16],[0,(w1 * ((r/rift_len)**n))[1600]],'--',color='0.5',linewidth=2)
ax1.plot([17,17],[0,(w1 * ((r/rift_len)**n))[1700]],'--',color='0.5',linewidth=2)
    
ax1.set_xlim(-5,45)
ax1.set_ylim(0,400)
ax1.grid('on')
    
ax1.annotate('GNSS w meas.: 254.5 ma$^{-1}$',(1,365),fontsize=ftsz)    
ax1.annotate('GNSS w model: '+str(round((w1 * ((r/rift_len)**n))[0],2))+' ma$^{-1}$',(1,335),fontsize=ftsz)
ax1.annotate('1km up rift:       '+str(round((w1 * ((r/rift_len)**n))[100],2))+' ma$^{-1}$',(1,305),fontsize=ftsz)
ax1.annotate('GNSS e meas.: 175.4 ma$^{-1}$',(21,365),fontsize=ftsz)
ax1.annotate('GNSS e model: '+str(round((w1 * ((r/rift_len)**n))[1600],2))+' ma$^{-1}$',(21,335),fontsize=ftsz)
ax1.annotate('1km up rift:      '+str(round((w1 * ((r/rift_len)**n))[1700],2))+' ma$^{-1}$',(21,305),fontsize=ftsz)  
    
ax1.set_ylabel('Rift width (m)')   
ax1.set_title('Rift width model (flow law n=1)',fontsize=ftsz)  

#-----------------------------------------------------
ax2=plt.subplot(3,2,3)

n=2

pl_t1 = ax2.plot(xx,(w1 * ((r/rift_len)**n)),'-',linewidth=2,color=(1,2/3,2/3),label='1 year')
    
ax2.plot([0,0],[0,w1],'--',color='0.5',linewidth=2)
ax2.plot([1,1],[0,(w1 * ((r/rift_len)**n))[100]],'--',color='0.5',linewidth=2)
    
#test fit at GNSS
ax2.plot([16,16],[0,(w1 * ((r/rift_len)**n))[1600]],'--',color='0.5',linewidth=2)
ax2.plot([17,17],[0,(w1 * ((r/rift_len)**n))[1700]],'--',color='0.5',linewidth=2)
    
ax2.set_xlim(-5,45)
ax2.set_ylim(0,400)
ax2.grid('on')
    
ax2.annotate('GNSS w meas.: 254.5 ma$^{-1}$',(1,365),fontsize=ftsz)    
ax2.annotate('GNSS w model: '+str(round((w1 * ((r/rift_len)**n))[0],2))+' ma$^{-1}$',(1,335),fontsize=ftsz)
ax2.annotate('1km up rift:       '+str(round((w1 * ((r/rift_len)**n))[100],2))+' ma$^{-1}$',(1,305),fontsize=ftsz)
ax2.annotate('GNSS e meas.: 175.4 ma$^{-1}$',(21,365),fontsize=ftsz)
ax2.annotate('GNSS e model: '+str(round((w1 * ((r/rift_len)**n))[1600],2))+' ma$^{-1}$',(21,335),fontsize=ftsz)
ax2.annotate('1km up rift:      '+str(round((w1 * ((r/rift_len)**n))[1700],2))+' ma$^{-1}$',(21,305),fontsize=ftsz)  

ax2.set_ylabel('Rift width (m)')
ax2.set_title('Rift width model (flow law n=2)',fontsize=ftsz)

#-----------------------------------------------------
ax3=plt.subplot(3,2,5)

n=3

pl_t1 = ax3.plot(xx,(w1 * ((r/rift_len)**n)),'-',linewidth=2,color=(1,2/3,2/3),label='1 year')
    
ax3.plot([0,0],[0,w1],'--',color='0.5',linewidth=2)
ax3.plot([1,1],[0,(w1 * ((r/rift_len)**n))[100]],'--',color='0.5',linewidth=2)
    
#test fit at GNSS
ax3.plot([16,16],[0,(w1 * ((r/rift_len)**n))[1600]],'--',color='0.5',linewidth=2)
ax3.plot([17,17],[0,(w1 * ((r/rift_len)**n))[1700]],'--',color='0.5',linewidth=2)
    
ax3.set_xlim(-5,45)
ax3.set_ylim(0,400)
ax3.grid('on')
    
ax3.annotate('GNSS w meas.: 254.5 ma$^{-1}$',(1,365),fontsize=ftsz)    
ax3.annotate('GNSS w model: '+str(round((w1 * ((r/rift_len)**n))[0],2))+' ma$^{-1}$',(1,335),fontsize=ftsz)
ax3.annotate('1km up rift:       '+str(round((w1 * ((r/rift_len)**n))[100],2))+' ma$^{-1}$',(1,305),fontsize=ftsz)
ax3.annotate('GNSS e meas.: 175.4 ma$^{-1}$',(21,365),fontsize=ftsz)
ax3.annotate('GNSS e model: '+str(round((w1 * ((r/rift_len)**n))[1600],2))+' ma$^{-1}$',(21,335),fontsize=ftsz)
ax3.annotate('1km up rift:      '+str(round((w1 * ((r/rift_len)**n))[1700],2))+' ma$^{-1}$',(21,305),fontsize=ftsz)  

ax3.set_ylabel('Rift width (m)')
ax3.set_xlabel('Distance along rift from western GNSS (km)')
ax3.set_title('Rift width model (flow law n=3)',fontsize=ftsz)

#-----------------------------------------------------
ax4=plt.subplot(3,2,2)
 
rift_len = 50
xx = np.arange(0,rift_len,0.01)
r = np.sqrt(rift_len**2 - xx**2)    
 
n=1
w1=259.75
   
pl_t1 = ax4.plot(xx,(w1 * ((r/rift_len)**n)),'-',linewidth=2,color=(1,2/3,2/3),label='1 year')
    
ax4.plot([10,10],[0,(w1 * ((r/rift_len)**n))[1100]],'--',color='0.5',linewidth=2)
ax4.plot([11,11],[0,(w1 * ((r/rift_len)**n))[1100]],'--',color='0.5',linewidth=2)
    
#test fit at GNSS
ax4.plot([26,26],[0,(w1 * ((r/rift_len)**n))[2600]],'--',color='0.5',linewidth=2)
ax4.plot([27,27],[0,(w1 * ((r/rift_len)**n))[2700]],'--',color='0.5',linewidth=2)
    
ax4.set_xlim(-5,55)
ax4.set_ylim(0,400)
ax4.grid('on')
    
ax4.annotate('GNSS w meas.: 254.5 ma$^{-1}$',(1,365),fontsize=ftsz)    
ax4.annotate('GNSS w model: '+str(round((w1 * ((r/rift_len)**n))[1000],2))+' ma$^{-1}$',(1,335),fontsize=ftsz)
ax4.annotate('1km up rift:       '+str(round((w1 * ((r/rift_len)**n))[1100],2))+' ma$^{-1}$',(1,305),fontsize=ftsz)
ax4.annotate('GNSS e meas.: 175.4 ma$^{-1}$',(31,365),fontsize=ftsz)
ax4.annotate('GNSS e model: '+str(round((w1 * ((r/rift_len)**n))[2600],2))+' ma$^{-1}$',(31,335),fontsize=ftsz)
ax4.annotate('1km up rift:      '+str(round((w1 * ((r/rift_len)**n))[2700],2))+' ma$^{-1}$',(31,305),fontsize=ftsz)  
    
ax4.set_title('Rift width model (flow law n=1)',fontsize=ftsz)    
    
#-----------------------------------------------------
ax5=plt.subplot(3,2,4)

n=2
w1=265.1
   
pl_t1 = ax5.plot(xx,(w1 * ((r/rift_len)**n)),'-',linewidth=2,color=(1,2/3,2/3),label='1 year')
    
ax5.plot([10,10],[0,(w1 * ((r/rift_len)**n))[1100]],'--',color='0.5',linewidth=2)
ax5.plot([11,11],[0,(w1 * ((r/rift_len)**n))[1100]],'--',color='0.5',linewidth=2)
    
#test fit at GNSS
ax5.plot([26,26],[0,(w1 * ((r/rift_len)**n))[2600]],'--',color='0.5',linewidth=2)
ax5.plot([27,27],[0,(w1 * ((r/rift_len)**n))[2700]],'--',color='0.5',linewidth=2)
    
ax5.set_xlim(-5,55)
ax5.set_ylim(0,400)
ax5.grid('on')
    
ax5.annotate('GNSS w meas.: 254.5 ma$^{-1}$',(1,365),fontsize=ftsz)    
ax5.annotate('GNSS w model: '+str(round((w1 * ((r/rift_len)**n))[1000],2))+' ma$^{-1}$',(1,335),fontsize=ftsz)
ax5.annotate('1km up rift:       '+str(round((w1 * ((r/rift_len)**n))[1100],2))+' ma$^{-1}$',(1,305),fontsize=ftsz)
ax5.annotate('GNSS e meas.: 175.4 ma$^{-1}$',(31,365),fontsize=ftsz)
ax5.annotate('GNSS e model: '+str(round((w1 * ((r/rift_len)**n))[2600],2))+' ma$^{-1}$',(31,335),fontsize=ftsz)
ax5.annotate('1km up rift:      '+str(round((w1 * ((r/rift_len)**n))[2700],2))+' ma$^{-1}$',(31,305),fontsize=ftsz)  



ax5.set_title('Rift width model (flow law n=2)',fontsize=ftsz)

#-----------------------------------------------------
ax6=plt.subplot(3,2,6)

n=3
w1=270.575
   
pl_t1 = ax6.plot(xx,(w1 * ((r/rift_len)**n)),'-',linewidth=2,color=(1,2/3,2/3),label='1 year')
    
ax6.plot([10,10],[0,(w1 * ((r/rift_len)**n))[1100]],'--',color='0.5',linewidth=2)
ax6.plot([11,11],[0,(w1 * ((r/rift_len)**n))[1100]],'--',color='0.5',linewidth=2)
    
#test fit at GNSS
ax6.plot([26,26],[0,(w1 * ((r/rift_len)**n))[2600]],'--',color='0.5',linewidth=2)
ax6.plot([27,27],[0,(w1 * ((r/rift_len)**n))[2700]],'--',color='0.5',linewidth=2)
    
ax6.set_xlim(-5,55)
ax6.set_ylim(0,400)
ax6.grid('on')

ax6.annotate('GNSS w meas.: 254.5 ma$^{-1}$',(1,365),fontsize=ftsz)    
ax6.annotate('GNSS w model: '+str(round((w1 * ((r/rift_len)**n))[1000],2))+' ma$^{-1}$',(1,335),fontsize=ftsz)
ax6.annotate('1km up rift:       '+str(round((w1 * ((r/rift_len)**n))[1100],2))+' ma$^{-1}$',(1,305),fontsize=ftsz)
ax6.annotate('GNSS e meas.: 175.4 ma$^{-1}$',(31,365),fontsize=ftsz)
ax6.annotate('GNSS e model: '+str(round((w1 * ((r/rift_len)**n))[2600],2))+' ma$^{-1}$',(31,335),fontsize=ftsz)
ax6.annotate('1km up rift:      '+str(round((w1 * ((r/rift_len)**n))[2700],2))+' ma$^{-1}$',(31,305),fontsize=ftsz)  

ax6.set_title('Rift width model (flow law n=3)',fontsize=ftsz)
ax6.set_xlabel('Distance along rift from MIR (km)')


ax1.annotate("a",(-5+2.5,400-40),ha='center',fontsize=16,fontweight='bold')
ax2.annotate("b",(-5+2.5,400-40),ha='center',fontsize=16,fontweight='bold')
ax3.annotate("c",(-5+2.5,400-40),ha='center',fontsize=16,fontweight='bold')
ax4.annotate("d",(-5+3,400-40),ha='center',fontsize=16,fontweight='bold')
ax5.annotate("e",(-5+3,400-40),ha='center',fontsize=16,fontweight='bold')
ax6.annotate("f",(-5+3,400-40),ha='center',fontsize=16,fontweight='bold')

output = figure_folder+'/figS15_upstream_1km.png'
plt.savefig(output, dpi=100, bbox_inches='tight')
plt.show()
```
